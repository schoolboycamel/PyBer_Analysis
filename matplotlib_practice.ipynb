```python
%matplotlib inline
import matplotlib.pyplot as plt
import statistics
import numpy as np 
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.plot(x_axis, y_axis, marker= '*', color = 'green', linewidth= 2.3, label='Boston')
plt.xlabel("dates")
plt.ylabel("Fare ($)")
plt.title("Byper Fare by Month")
plt.grid()
plt.ylim(0, 45)
plt.xlim(0, 11.5)
plt.legend()
```

#CREATE VERTICAL BAR CHART 

```python
plt.bar(x_axis, y_axis, color= "green", label= 'Boston')
plt.ylim(0,45)
plt.xlim(0,12.5)
plt.xlabel("Dates")
plt.ylabel('Fares($)')
plt.title('Byper Fares per Month')
plt.legend()
```

#CREATE HORIZONTAL BAR CHART

```python
plt.barh(x_axis, y_axis, color= "pink", label= "Boston")
plt.xlabel('Fares ($)')
plt.ylabel('Date')
plt.title('ByPer Fares per Month')
plt.gca().invert_yaxis()
```

#SCATTER PLOTS 

```python
plt.scatter(y_axis,x_axis, color= 'red', label= 'Chicago')
plt.xlabel('Fare ($)')
plt.ylabel('Dates')
plt.legend()
plt.gca().invert_yaxis()
```

#BUBBLE PLOTS

```python
plt.scatter(x_axis, y_axis, s=y_axis)
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

    #WITHOUT USING FOR LOOP!

```python
plt.scatter(y_axis, x_axis, color= "blue", s = [i * 5 for i in y_axis], label= "Quito")
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Dates")
plt.legend()
```

#PIE CHART

```python
plt.pie(y_axis, labels=x_axis)
```

```python
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors= ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize=(10,10))
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct='%.1f%%')
```

#ERROR BARS TO LINE CHART: Adding error bars can show either the standard deviation, standard error, confidence intervals, or minimum and maximum values of a dataset. When added to a chart, they can visually show the variability of the plotted data. By looking at the error bars, one can infer the significance of the data.
plt.errorbar(x, y, yerr=<value>)

```python
#find standard deviation
stdev= statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize= 3, color= "blue")
```

#Error bars (BAR CHART)

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize= 3)
```

#CHANGE MAJOR TICKS 
X TICK 

```python
#create horizontal bar graph
plt.barh(x_axis, y_axis)
#plot ticks on x axis
plt.xticks(np.arange(0,51, step=5.0))
#invert y axis
plt.gca().invert_yaxis()
```

#MINOR TICKS 

```python
from matplotlib.ticker import MultipleLocator
# Increase the size of the plot figure.
fig, ax= plt.subplots(figsize=(8,8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
